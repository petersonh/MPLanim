```python
import matplotlib.pyplot as plt
import numpy as np
import math
import easing_functions as ef
from MPLanim import Timeline, anims
```

```python
fig, ax = plt.subplots(1)
steps = 5000

sin_x = [x for x in range(steps)]
sin_y = [math.sin(x/steps * math.pi * 6) for x in range(steps)]
sin_line_plot, = ax.plot([], [], color="blue", linewidth=3)
sin_dot_plot, = ax.plot([],[], marker='o', color="blue", markersize=10, markeredgecolor="white", markeredgewidth=2)

cos_x = [x for x in range(steps*2)]
cos_y = [math.cos(x/steps * math.pi * 6) for x in range(steps*2)]
cos_line_plot, = ax.plot([], [], color="red", linewidth=3)
cos_dot_plot, = ax.plot([],[], marker='o', color="red", markersize=10, markeredgecolor="white", markeredgewidth=2)
```

```python
x = np.linspace(0,10*math.pi, 5000)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x,y)
```

```python
tl = Timeline(fig, 30)

tl.add_anim(anims.LineDraw2(line=sin_line_plot, x=sin_x, y=sin_y, start=0, duration=2000))
tl.add_anim(anims.Autoscale2(start=0, duration=2000, ax=ax))
# tl.add_anim(anims.Maxscale2(start=0, duration=2000, ax=ax, x=sin_x, y=sin_y))
tl.add_anim(anims.MoveMarker2(start=0, duration=2000, marker=sin_dot_plot, x=sin_x, y=sin_y))
tl.add_anim(anims.ScaleX2(start=2000, duration=2000, ax=ax, left=0, right=3000))

        
#     def __init__(self, 
#                  start: int,
#                  duration: int,
#                  marker: matplotlib.lines.Line2D,
#                  x: list,
#                  y: list,
#                  easing: Callable[[float, float, float], float]):

tl.show_in_notebook()
```

```python
animations = []

annotation = ax.annotate("Flu 2009",
                            xy=(0,0), xycoords='data',
                            xytext=(0, 0), textcoords='data',
                            arrowprops=dict(arrowstyle="-", color='blue', linestyle='--'),
                            ha="center",
                            bbox=dict(facecolor='white', edgecolor='blue', boxstyle="round,pad=0.5"))
animations.append(annotation)

sin_line_draw = anims.LineDraw(start_time=0, 
                         duration=1000, 
                         easing_function=ef.SineEaseOut, 
                         line=sin_line_plot, 
                         x_data=sin_x, 
                         y_data=sin_y, 
                         start_index=0, 
                         end_index=(len(sin_x) - 1))
animations.append(sin_line_draw)

sin_dot_move = anims.MoveMarker(start_time=0, 
                          duration=1000, 
                          easing_function=ef.SineEaseOut, 
                          marker=sin_dot_plot, 
                          x_data=sin_x, 
                          y_data=sin_y, 
                          start_index=0, 
                          end_index=(len(sin_x) - 1))
animations.append(sin_dot_move)

cos_line_draw = anims.LineDraw(start_time=500, 
                         duration=1000, 
                         easing_function=ef.SineEaseOut, 
                         line=cos_line_plot, 
                         x_data=cos_x,
                         y_data=cos_y,
                         start_index=0,
                         end_index=(len(cos_x) - 1))
animations.append(cos_line_draw)

cos_dot_move = anims.MoveMarker(start_time=500, 
                          duration=1000, 
                          easing_function=ef.SineEaseOut, 
                          marker=cos_dot_plot, 
                          x_data=cos_x, 
                          y_data=cos_y, 
                          start_index=0, 
                          end_index=(len(cos_x) - 1))
animations.append(cos_dot_move)

autoscale = anims.Autoscale(start_time=0, 
                                  duration=2500, 
                                  ax=ax)
animations.append(autoscale)

scalex = anims.ScaleX(start_time=2500, 
                duration=1000, 
                ax=ax,
                left=4000,
                right=6000)
animations.append(scalex)

scaley = anims.ScaleY(start_time=2500, 
                duration=1000, 
                ax=ax,
                bottom=-0.25,
                top=0.25)
animations.append(scaley)

moving_marker = anims.AnnotateMovingMarker(start_time=0, 
                                     duration=1000, 
                                     easing_function=ef.SineEaseOut, 
                                     ax=ax,
                                     annotation=annotation,
                                     x_data=sin_x,
                                     y_data=sin_y,
                                     start_index=0, 
                                     end_index=(len(sin_x) - 1),
                                     y_text_position_pixels=250)
animations.append(moving_marker)
```

```python
animation = Timeline(animations, fig, 30)
animation.show_in_notebook()
# animation.save("test_output")
```
